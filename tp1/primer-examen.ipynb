```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga de los datos
df_ventas = pd.read_excel('./Datos_primer_TP_22Co2024_a2221.xlsx')
print("Columnas: ", df_ventas.columns)
print("Cantidad de datos: ", len(df_ventas))
print(df_ventas.head())
```

```python
def plot_functions(year: int):
    """
        Genera gráficos de la función empírica de distribución (ECDF) y la densidad (KDE) para un año específico.
    """
    # Asegurarse de que la columna "Fecha
    df_ventas_anio = df_ventas[df_ventas["Fecha"].dt.year == year]

    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Gráfico de la Función empírica de distribución (ECDF)
    sns.ecdfplot(data=df_ventas_anio, x="Ventas", linewidth=2, ax=axes[0])
    axes[0].set_title(f"Función empírica de distribución (ECDF) - Año {year}")
    axes[0].set_xlabel("Monto de Venta")
    axes[0].set_ylabel("F(x)")
    axes[0].grid(True, alpha=0.3)

    # Aproximacion a la funcion de densidad (KDE)
    sns.kdeplot(data=df_ventas_anio, x="Ventas", fill=True, ax=axes[1], linewidth=2)
    axes[1].set_title(f"Aproximacion a la funcion de densidad (KDE) - Año {year}")
    axes[1].set_xlabel("Monto de Venta")
    axes[1].set_ylabel("f(x)")
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
plot_functions(2021)
```

```python
plot_functions(2022)
```

```python
plot_functions(2023)
```
